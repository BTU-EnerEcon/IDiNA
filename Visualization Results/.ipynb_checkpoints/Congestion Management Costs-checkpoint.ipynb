```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


cm_to_inch = 0.393701
fig_width = 16 * cm_to_inch
fig_height = 10 * cm_to_inch

# Daten aus der Excel-Datei einlesen
df = pd.read_excel('Load Forecast Accuracy.xlsx')

# Busspalte in kategorialen Datentyp umwandeln
#df['Bus'] = pd.Categorical(df['Bus'], categories=df['Bus'].unique(), ordered=True)


plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 9,
    "axes.titlesize": 9,
    "axes.labelsize": 9,
    "text.color": "black",
    "legend.title_fontsize": 9
})

# Zeile mit "Average" entfernen und den Scatterplot erstellen
df_temp = df[df['Bus'] != 'Average']
plt.figure(figsize=(fig_width, fig_height))

sns.set_theme(style='white',font='Times New Roman', font_scale = 0.8)

sns.scatterplot(x='Bus', y='n RMSE SLP', data=df_temp, label='SLP',s=25, marker='s',color ='black')
sns.scatterplot(x='Bus', y='n RMSE ARIMA', data=df_temp, label='ARIMA',s=25,marker='x',color ='black')

last_data_point = df_temp.iloc[-1]  # Den letzten Datenpunkt auswählen
plt.scatter(34, last_data_point['n RMSE SLP'], marker='s', c='none', edgecolors='red', linewidths=2, s=100)
plt.scatter(34, last_data_point['n RMSE ARIMA'], marker='s', c='none', edgecolors='green', linewidths=2, s=100)

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Node')
plt.ylabel('Standard Deviation of the Forecast Error')
#plt.title('Forecast Method Comparison - Nodal SD')

# Legende anzeigen
plt.legend(title_fontsize = 9)


x_ticks = list(range(2,35))
x_tick_labels = df_temp['Bus'].tolist()
x_tick_labels = x_tick_labels[1:]
x_tick_labels[-1]='Average'

# Zeile wieder hinzufügen und sortieren
df_temp = pd.concat([df_temp, df[df['Bus'] == 'Average']])
df_temp = df_temp.sort_values(by='Bus')


#  x-Tick-Label um x Grad drehen
#x_ticks = x_ticks[1:]
plt.xticks(x_ticks, x_tick_labels, rotation=45, ha='right')

plt.tight_layout() 
# Diagramm als PDF speichern
#plt.savefig('Nodal_Forecast_Error.pdf')
plt.savefig('Nodal_SD.svg')
# Diagramm anzeigen
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cm_to_inch = 0.393701
fig_width = 16 * cm_to_inch
fig_height = 10 * cm_to_inch


plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 9,
    "axes.titlesize": 9,
    "axes.labelsize": 9,
    "text.color": "black",
    "legend.title_fontsize": 9
})

# Daten aus der Excel-Datei einlesen
df = pd.read_excel('Uniform_Rollout.xlsx')

# Seaborn-Plot für jedes Szenario erstellen
plt.figure(figsize=(fig_width , fig_height))

# Umwandlung der "Smart Meter"-Werte in Strings
df['Smart Meter'] = df['Smart Meter'].astype(str)

#Farben der Säulen festlegen
Farben = sns.cubehelix_palette(rot=-.2)

# Erste Farbe aus der Farbpalette extrahieren
StartFarbe = Farben[0]


ax1 = sns.barplot(x='Smart Meter', y='Congestion Management Costs', data=df, color = StartFarbe ,label='Costs')
ax1.set_ylabel('Relative Costs [%]')
#,hatch='/'

# Sekundäre y-Achse erstellen und Linien für Prognosegenauigkeit darauf darstellen (für jedes Szenario separat)
ax2 = ax1.twinx()
sns.lineplot(x='Smart Meter', y='Accuracy', data=df,ax=ax2, marker='s',color='black', label='Forecast Error')
ax2.set_ylabel('Forecast Error - Standard Deviation')
# linestyle='dashed',
# Achsenbeschriftungen und Titel hinzufügen
ax1.set_xlabel('Rollout Level in [%]')
plt.title('Congestion Management Costs and Forecast Accuracy')

# Legenden-Handles und -Labels von beiden Achsen kombinieren
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines += lines2
labels += labels2

# Legende anzeigen
plt.legend(lines, labels, loc='upper right', facecolor='white',framealpha=1)

# Diagramm als PDF speichern
#plt.savefig('Uniform_Rollout.pdf')
plt.savefig('Uniform_Rollout.svg')
# Diagramm anzeigen
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D

cm_to_inch = 0.393701
fig_width = 16 * cm_to_inch
fig_height = 10 * cm_to_inch


plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 9,
    "axes.titlesize": 9,
    "axes.labelsize": 9,
    "text.color": "black",
    "legend.title_fontsize": 9
})

# Daten aus der Excel-Datei einlesen
df = pd.read_excel('Strategic_Rollout_v2.xlsx')

# Umwandlung der "Smart Meter"-Werte in Strings
df['Smart Meter'] = df['Smart Meter'].astype(str)

# Seaborn-Plot erstellen
plt.figure(figsize=(fig_width, fig_height))

#Farben der Säulen festlegen
Farben = sns.cubehelix_palette(rot=-.2)

# Schraffur-Muster für die Szenarien
#hatch_patterns = {'Uniform Rollout': '/', 'Strategic Rollout ': '-'}

# Linien für "Relative Costs" auf primärer y-Achse darstellen (Säulendiagramm)
ax1 = sns.barplot(x='Smart Meter', y='Relative Costs', hue='Rollout', data=df, palette=Farben)

# Schraffur für jedes Szenario festlegen
#for bar, scenario in zip(ax1.patches, df['Scenario']):
#    hatch = hatch_patterns.get(scenario)
#    if hatch:
#        bar.set_hatch(hatch)
                  
ax1.set_ylabel('Relative Costs [%]')

# Daten für "Accuracy" auf sekundärer y-Achse darstellen (Liniendiagramm)
ax2 = ax1.twinx()
sns.lineplot(x='Smart Meter', y='Accuracy', style='Rollout', color='black',markers=True , data=df, ax=ax2, legend=False)
ax2.set_ylabel('Forecast Error - Standard Deviation')

# Achsenbeschriftungen und Titel hinzufügen
ax1.set_xlabel('Rollout Level in [%]')
plt.title('Comparison: Uniform Rollout vs. Strategic Rollout')

# Legende für Linienstil manuell erstellen
line_styles = ['-', ':', '--', ]  # Linienstile für die Szenarien
legend_lines = [plt.Line2D([0], [0], color='black', linestyle=style) for style in line_styles]

# Angepasste Legende erstellen
legend_labels = ['Costs Uniform', 'Costs Strategic', 'Forecast Error Uniform', 'Forecast Error Strategic']
legend_elements = [plt.Line2D([0], [0], color=Farben[0], lw=4, label='Costs Uniform'),
                   plt.Line2D([0], [0], color=Farben[1], lw=4, label='Costs Strategic'),
                   plt.Line2D([0], [0], color='black', linestyle='-', lw=2, label='Forecast Error Uniform'),
                   plt.Line2D([0], [0], color='black', linestyle=':', lw=2, label='Forecast Error Strategic')]


# Legende außerhalb des Diagramms platzieren
ax1.legend(handles=legend_elements, loc='upper right', title='Rollout')

# Legende außerhalb des Diagramms platzieren
#ax1.legend([*ax1.get_legend().legendHandles, *legend_lines], [*df['Rollout'].unique(), *df['Rollout'].unique()], loc='upper right', title='Rollout',)

# Diagramm als PDF speichern
#plt.savefig('Rollout_Scenario_Comparison.pdf')
plt.savefig('Rollout_Scenario_Comparison.svg')

# Diagramm anzeigen
plt.show()


```
