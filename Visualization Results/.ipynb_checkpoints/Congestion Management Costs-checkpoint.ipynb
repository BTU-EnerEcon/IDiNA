```python
import pandas as pd
import seaborn as sns

sns.set_theme(style="dark", font_scale=1.5)
costs = pd.read_excel("Seaborn_input.xlsx")

# Plot each year's time series in its own facet
g = sns.relplot(
    data=costs,
    x="S", y="Values", col="Case", hue="Case",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=2, height=4, aspect=1.5, legend=False,
)

# Iterate over each subplot to customize further
for Case, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, Case, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=costs, x="S", y="Values", units="Case",
        estimator=None, color=".7", linewidth=2, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::200])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Costs in EUR")
g.tight_layout()

```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
# Load dataset with long-form data
RD_costs = pd.read_excel("Seaborn_robust_redispatch.xlsx")
RD_costs['Scenarios'] = pd.Categorical(RD_costs['Scenarios'], categories=['Sc 1', 'Sc 2', 'Sc 3', 'Sc 4', 'Sc 5', 'Sc 6', 'Sc 7', 'Sc 8', 'Sc 9', 'Sc 10'])
RD_costs=RD_costs.pivot(index="Scenarios", columns="Season", values="Costs")


# Plot the responses for different events and regions
sns.lineplot(data=RD_costs, palette ="rocket",linewidth = 3)

sns.set(rc={"figure.figsize":(8, 5)})
#plt.xlabel("Season", size=16)
plt.ylabel("Cost in % of Total System Costs")
```
