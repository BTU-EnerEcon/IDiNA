```python
import pandas as pd
import numpy as np

# Datei lesen
df = pd.read_excel('combined_df_2021.xlsx')
print(df.head())
```

```python
# Standardize column names by renaming them sequencially
column_number = 0
for column in df.columns:
    df.rename(columns={column: 'col'+str(column_number)}, inplace=True)
    column_number += 1
```

```python
df.head().style \
  .format(precision=3, thousands=".", decimal=",") \
  .format_index(str.upper, axis=1)
```

```python
list(df.columns)
```

```python
import warnings

warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
```

```python
max_value = [100, 90, 120, 60, 60, 200, 200, 60, 60, 45, 60, 60, 120, 60, 60, 60, 90, 90, 90, 90, 90, 90, 420, 420, 60, 60, 60, 120, 200, 150, 210, 60]
max_value_index = 0
result = pd.DataFrame()

# Create an empty dataframe
temp_df = pd.DataFrame()
for column in df.columns:
    temp_sum = temp_df.sum(axis=1)
    if (temp_sum < max_value[max_value_index]).all():
        temp_df[column] = df[column]
    elif (temp_sum > max_value[max_value_index]).any():
        # Drop the last column
        temp_df = temp_df.iloc[:, :-1]
        result["new"+str(max_value_index)] = temp_df.sum(axis=1)
        print("The column new"+str(max_value_index)+" is created from the following columns:")
        print(temp_df.columns)
        temp_df = pd.DataFrame()
        if max_value_index < len(max_value)-1:
            max_value_index += 1
        else:
            break
    else:
        result["new"+str(max_value_index)] = temp_df.sum(axis=1)
        temp_df = pd.DataFrame()
        if max_value_index < len(max_value)-1:
            max_value_index += 1
        else:
            break
```

```python
result.head().style \
  .format(precision=3, thousands=".", decimal=",") \
  .format_index(str.upper, axis=1)
```

```python
result.to_excel('test_output_Phong.xlsx', index=False)
```
